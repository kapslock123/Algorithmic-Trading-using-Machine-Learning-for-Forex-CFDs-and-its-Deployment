#Signal Generation

This line helps to set a prerequisite to trade. 4 more features are added which stores signals of individual indicators. **Please note that this values doesn't determine our position taken in the market, since it will be decided by the model only.**

image.png

```python
def Signal_1(Bid1):
  Bid1['Signal_B']=0
  Bid1['Signal_SMA']=0
  Bid1['Signal_A']=0
  Bid1['Signal_R']=0
  Bid1['Signal_M']=0
  Bid1.loc[Bid1.BB_U<=Bid1.c,'Signal_B']=-1 #if price crosses upper bollinger band indicate sell
  Bid1.loc[Bid1.BB_L>=Bid1.c,'Signal_B']=1 #if price crosses lower bollinger band indicate buy
  Bid1['Signal_B']=Bid1.Signal_B.map(lambda x: x if (x==1 or x==-1)  else 0)  # replace NaN values with 0
  Bid1.loc[((Bid1['SMA_10'].shift(1)<Bid1['SMA_30'].shift(1)) & (Bid1['SMA_10']>Bid1['SMA_30'])), 'Signal_SMA' ]=1  # if fast Simple moving average crosses slow simple moving average input buy or sell based on direction of crossing
  Bid1.loc[((Bid1['SMA_10'].shift(1)>Bid1['SMA_30'].shift(1)) & (Bid1['SMA_10']<Bid1['SMA_30'])), 'Signal_SMA' ]=-1
  
  Bid1['Signal_SMA']=Bid1.Signal_SMA.map(lambda x: x if (x==1 or x==-1) else 0) # replace NaN values with 0
  Bid1.loc[(Bid1.Rsi>=70),'Signal_R']=-1 # If RSI greater than 70 sell signal.
  Bid1.loc[(Bid1.Rsi<=30),'Signal_R']=1  
  
  Bid1['Signal_R']=Bid1.Signal_R.map(lambda x: x if (x==1 or x==-1) else 0) # replace NaN values with 0
  
  Bid1.loc[(Bid1.ADX>=25) & (Bid1.ADX_pos<=Bid1.ADX_neg),'Signal_A']=-1 # if ADX>25 and +DI<-DI (ADX lines.), sell signal otherwise buy.

  Bid1.loc[(Bid1.ADX>=25) & (Bid1.ADX_pos>=Bid1.ADX_neg),'Signal_A']=1
  Bid1['Signal_A']=Bid1.Signal_A.map(lambda x: x if (x==1 or x==-1) else 0) # replace NaN values with 0

  Bid1.loc[(Bid1['MACD'].shift(1)<Bid1['MACD_Signal'].shift(1)) & (Bid1['MACD']>Bid1['MACD_Signal']), 'Signal_M' ]=1 #MACD signal condition
  
  Bid1.loc[(Bid1['MACD'].shift(1)>Bid1['MACD_Signal'].shift(1)) & (Bid1['MACD']<Bid1['MACD_Signal']), 'Signal_M' ]=-1
  Bid1['Signal_M']=Bid1.Signal_M.map(lambda x: x if (x==1 or x==-1) else 0) # replace NaN values with 0




```

#Generating training Dataset

To understand it in detail, let’s consider a stock price of 100. The price of this stock may go up, down and sideways. It can reach a price of 120 without its value going below 100 and can reach 120 after it has gone down to 80. In both the cases the price of 120 is reached but if the trader has set a stop loss of 85 considering a buy position in the market, the trader’s trade will close at 85 without reaching 120 and since the trade has incurred a loss with the buy position the correct label for that data point should’ve been sell instead of buy. From this we can understand that the classes of the training dataset are sure to change for a financial time series data based on the target profit and stop loss of a trader. If we were to label each and every data point in the dataset according to the target profit and stop loss set by the trader, each and every data point will be assigned a class which would either be buy or sell. Hence a pre-requisite has to be set for when to initiate a trade be it buy or sell and then feed the dataset to the model for training.

For the first training dataset, the prerequisite for trading is set such that a position will be created only when at least 3 indicators show a signal. 

```python
def train_dataset1(Bid1,aprofit):
  x=Bid1.loc[((abs(Bid1.Signal_M)+abs(Bid1.Signal_R)+abs(Bid1.Signal_A)+abs(Bid1.Signal_SMA)+abs(Bid1.Signal_B))>=3 )].index
  
  index_count=0
  store_index=0
  profit=[]
  loss=[]
  while(True):
    count=0
    
    if index_count==len(x)-1:
      break
    print(index_count)
    '''
    For finding out if after the point at which our condition is met, did the price hit target profit first or stop loss first,
    we have to traverse through the dataset from the point the position is created to the point it reached either target profit or stop loss
    From the next line traversing through original dataset is done. 
    '''
    for i in Bid1.loc[x[index_count]:].iterrows():
      
      if count==0:
        differentialcount=i[1][6]-i[1][7]
        profit.append(ProfitAgainst(differentialcount))
        loss.append(MaxLoss(differentialcount))
        close=i[1][5]
        count+=1
      # print((i[1][5]-close)*3000)
      
      if (i[1][5]-close)*3000*aprofit>=ProfitAgainst(differentialcount):
        
        Bid1.loc[x[index_count],'Signal']=1
        store_index=index_count
        index_count+=1
        break



      if (i[1][5]-close)*3000*aprofit<=MaxLoss(differentialcount)*2:
        
        Bid1.loc[x[index_count],'Signal']=-1
        store_index=index_count
        index_count+=1
        
        break

      '''
      sometimes, the datapoint which follows the prerequisite is at the very end and there is no data to traverse hence we put nan value in signal column.
      '''
      if (i[0]==Bid1.iloc[[-1]].index):
        
        Bid1.loc[x[index_count],'Signal']=np.NaN
        store_index=index_count
        index_count+=1
        
      '''

      The new signal column will give the actual label to the dataset.

      '''
  return Bid1.loc[x],profit,loss



      

```

**For training dataset 2**

Before moving ahead with the model it is necessary to calculate the hyperparameters. Machine learning algorithms feature adjustable parameters also known as the hyperparameters. In order to develop strong, accurate and efficient models, before training the model setting the hyperparameters values are extremely important. Since there is always a tradeoff between bias and variance, hyperparameters are used to assist us in the tradeoff which inturn prevents underfitting and overfitting of the models. 
In case of SVM with rbf kernal, c and gamma are the hyper parameters.

**Hyperparameter tunning for each currency pairs**

```python
def SVM_Model_hyperparameters(pair, hyperparams,label):
  Bid1=pair.copy()
  
  split_factor = .8
  split = int(split_factor*len(Bid1))
  

  '''
  It is necessary to scale the values in each features before fitting in the model in case of SVM for this we use pipelines which is nothing but an object containing steps to 
  be performed before finding the hyper parameters.
  '''
  pipeline_steps = [('scaler', StandardScaler()), ('svc', SVC())]
  
  # we shall pass the step inside the pipeline object
  pipeline= Pipeline(pipeline_steps)
  
  hyperparameters = {'svc__C': [10, 100, 1000, 10000], 'svc__gamma': [1e-2, 1e-1, 1e0], 'svc__kernel': ['rbf']}
  
  clf = RandomizedSearchCV(pipeline, hyperparameters, cv=TimeSeriesSplit(n_splits=2))
  y=Bid1['Signal']
  X=Bid1.drop(['h','l','c','o','complete','volume','DayHigh','DayLow','BB_L','BB_U','Signal'],axis=1)
  # We shall fit our data and let the randomizedsearch cv perform the operation of finding best parameters
  clf.fit(X.iloc[:split], y.iloc[:split])

  '''
  Next step is to store the best parameters to a variable
  '''
  C = clf.best_params_['svc__C']

  kernel = clf.best_params_['svc__kernel']

  gamma = clf.best_params_['svc__gamma']
  hyperparams[label].append(C)
  hyperparams[label].append(kernel)
  hyperparams[label].append(gamma)
  
  

  
```

```python
Pairs=[B1,B2,B3,B4,B5,B6,B7,B8,B9,B10,B11,B12,B13]
SVM_hyperparameters={"B1":[],"B2":[],"B3":[],"B4":[],"B5":[],"B6":[],"B7":[],"B8":[],"B9":[],"B10":[],"B11":[],"B12":[],"B13":[]}
count=0
label=["B1","B2","B3","B4","B5","B6","B7","B8","B9","B10","B11","B12","B13"]
for i in Pairs:
    SVM_Model_hyperparameters(i,SVM_hyperparameters,label[count])
    count+=1
    
pd.DataFrame(SVM_hyperparameters,index=['svc__C',"Kernel","Gamma"])
```

```python
def SVM_Model_iterative(pair,hyperparameter,label,cr,cm):
  Bid1=pair.copy()
  
  split_factor = .8
  split = int(split_factor*len(Bid1))
  y=Bid1['Signal']
  X=Bid1.drop(['h','l','c','o','complete','volume','DayHigh','DayLow','BB_L','BB_U','Signal'],axis=1)
  
  C = hyperparameter[label][0]
  kernel = hyperparameter[label][1]
  gamma = hyperparameter[label][2]
  clf = SVC(C=C, kernel=kernel, gamma=gamma)
  scaler= StandardScaler()
  '''
  Now that we have the best parameters we shall fit into the SVC()
  '''

  clf.fit(scaler.fit_transform(X.iloc[:split]), y.iloc[:split])

  y_predict = clf.predict(scaler.transform(X.iloc[split:]))
  
  Bid1['y_predict'] = 0
  store = pd.Series( clf.predict( scaler.transform(X.iloc[:split])).tolist() )
  Bid1.iloc[:split,Bid1.columns.get_loc('y_predict')] = store

  
  Bid1.iloc[split:,Bid1.columns.get_loc('y_predict')] = y_predict

  # confusion matrix
  cm_i = confusion_matrix(y[split:], y_predict)
  # classification report
  cr_i = classification_report(y[split:], y_predict,output_dict=True)
#   print(cr)
  cr.append(cr_i)
  cm.append(cm_i)
```

It is obvious that since trading happens throughout the day, some trades will stay open after interruption because they haven't reached their target profit and stop losses. so we shall calculate the accuracy for those trades which are closed. so we shall only consider open positions till the length of signal

```python
def Deployement_Database(Biddd1,label):
  kapil=Biddd1.copy()
  Signal=kapil.loc[(kapil.Signal>0) | (kapil.Signal<0) ,'Signal']
  y_predict=kapil.loc[(kapil.Position>0) | (kapil.Position<0), 'Position' ][:len(Signal)]
  Target_profit=kapil.loc[(kapil.Position>0) | (kapil.Position<0), 'Target_Profit' ][:len(Signal)]
  Stop_loss=kapil.loc[(kapil.Position>0) | (kapil.Position<0), 'Stop_Loss' ][:len(Signal)]
  df=pd.DataFrame({"label":np.repeat(label,len(y_predict)),"Predicted_Signal": y_predict.to_list(),"Actual_signal": Signal.to_list(),"Target_Profit": Target_profit.to_list(),"Stop_Loss":Stop_loss.to_list(),"Trade_Open_time":y_predict.index,"Trade_close_time":Signal.index})
  df.loc[(df.Actual_signal==df.Predicted_Signal),'P/L']=df.loc[(df.Actual_signal==df.Predicted_Signal),'Target_Profit']
  df.loc[(df.Actual_signal!=df.Predicted_Signal),'P/L']=df.loc[(df.Actual_signal!=df.Predicted_Signal),'Stop_Loss']
  return df

```

I have added target profit and stop loss for each trade when the trade was opened. so below is the calculation of total profit. The condition is if current profit>target profit close the trade, so there may be current profits which are greater than target profit but I have neglected the minor difference since accuracy was my performance measure 

This model takes every 10 minute candle. to calculate it from tick prices we resample the last bar in another variable and update the price every 10 minutes. since I am only trading with 0.03 lots or 3000 units the overall profits is approximately 94.77 GBP for an investment of around 1000 GBP which involves some part in margin money and around 200 GBP as buffer amount.
