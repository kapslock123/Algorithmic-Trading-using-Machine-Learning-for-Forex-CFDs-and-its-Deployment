```python
def ADX(Bid1):
  Bid1['ADX'],Bid1['ADX_pos'],Bid1['ADX_neg']=ta.trend.adx(Bid1.h,Bid1.l,Bid1.c),ta.trend.adx_pos(Bid1.h,Bid1.l,Bid1.c),ta.trend.adx_neg(Bid1.h,Bid1.l,Bid1.c)
```

```python
def RSI(Bid1):  
  Bid1['Rsi']=ta.momentum.rsi(Bid1.c)
```

```python
def MACD(Bid1):
  Bid1['MACD'],Bid1['MACD_Signal'],Bid1['MACD_hist']=ta.trend.MACD(Bid1.c).macd(),ta.trend.MACD(Bid1.c).macd_signal(),ta.trend.MACD(Bid1.c).macd_diff()
```

```python
def MA(Bid1):
  Bid1['SMA_10'],Bid1['SMA_30']=Bid1['c'].rolling(window=10).mean(),Bid1['c'].rolling(window=30).mean()
```

```python
def BBands(Bid1):
  Bid1['BB_U'],Bid1['BB_L']=ta.volatility.bollinger_hband(Bid1.c,window_dev=2.3),ta.volatility.bollinger_lband(Bid1.c,window_dev=2.3)
```

```python
def Indicators(Bid1):
  RSI(Bid1)
  MACD(Bid1)
  ADX(Bid1)
  MA(Bid1)
  BBands(Bid1)
  
```

#Signal Generation

```python
def Signal_1(Bid1):
  Bid1['Signal_B']=0
  Bid1['Signal_SMA']=0
  Bid1['Signal_A']=0
  Bid1['Signal_R']=0
  Bid1['Signal_M']=0
  Bid1.loc[Bid1.BB_U<=Bid1.c,'Signal_B']=-1
  Bid1.loc[Bid1.BB_L>=Bid1.c,'Signal_B']=1
  Bid1['Signal_B']=Bid1.Signal_B.map(lambda x: x if (x==1 or x==-1)  else 0)
  Bid1.loc[((Bid1['SMA_10'].shift(1)<Bid1['SMA_30'].shift(1)) & (Bid1['SMA_10']>Bid1['SMA_30'])), 'Signal_SMA' ]=1
  Bid1.loc[((Bid1['SMA_10'].shift(1)>Bid1['SMA_30'].shift(1)) & (Bid1['SMA_10']<Bid1['SMA_30'])), 'Signal_SMA' ]=-1
  
  Bid1['Signal_SMA']=Bid1.Signal_SMA.map(lambda x: x if (x==1 or x==-1) else 0)
  Bid1.loc[(Bid1.Rsi>=70),'Signal_R']=-1
  Bid1.loc[(Bid1.Rsi<=30),'Signal_R']=1
  
  Bid1['Signal_R']=Bid1.Signal_R.map(lambda x: x if (x==1 or x==-1) else 0)
  
  Bid1.loc[(Bid1.ADX>=25) & (Bid1.ADX_pos<=Bid1.ADX_neg),'Signal_A']=-1

  Bid1.loc[(Bid1.ADX>=25) & (Bid1.ADX_pos>=Bid1.ADX_neg),'Signal_A']=1
  Bid1['Signal_A']=Bid1.Signal_A.map(lambda x: x if (x==1 or x==-1) else 0)

  Bid1.loc[(Bid1['MACD'].shift(1)<Bid1['MACD_Signal'].shift(1)) & (Bid1['MACD']>Bid1['MACD_Signal']), 'Signal_M' ]=1
  
  Bid1.loc[(Bid1['MACD'].shift(1)>Bid1['MACD_Signal'].shift(1)) & (Bid1['MACD']<Bid1['MACD_Signal']), 'Signal_M' ]=-1
  Bid1['Signal_M']=Bid1.Signal_M.map(lambda x: x if (x==1 or x==-1) else 0)




```

```python
def train_dataset(Bid1,aprofit):
  x=Bid1.loc[(abs(Bid1.Signal_M)+abs(Bid1.Signal_R)+abs(Bid1.Signal_A)+abs(Bid1.Signal_SMA)+abs(Bid1.Signal_B))>=3].index
  
  index_count=0
  store_index=0
  profit=[]
  loss=[]
  while(True):
    count=0
    
    if index_count==len(x)-1:
      break
    print(index_count)
    for i in Bid1.loc[x[index_count]:].iterrows():
      # print(i[0] == x[index_count])
      
      # print(i)
      if count==0:
        differentialcount=i[1][6]-i[1][7]
        profit.append(ProfitAgainst(differentialcount))
        loss.append(MaxLoss(differentialcount))
        close=i[1][5]
        count+=1
      # print((i[1][5]-close)*3000)
      
      if (i[1][5]-close)*3000*aprofit>=ProfitAgainst(differentialcount):
        
        Bid1.loc[x[index_count],'Signal']=1
        store_index=index_count
        index_count+=1
        break



      if (i[1][5]-close)*3000*aprofit<=MaxLoss(differentialcount)*4:
        
        Bid1.loc[x[index_count],'Signal']=-1
        store_index=index_count
        index_count+=1
        
        break
      if (i[0]==Bid1.iloc[[-1]].index):
        
        Bid1.loc[x[index_count],'Signal']=np.NaN
        store_index=index_count
        index_count+=1
        
  return Bid1.loc[x],profit,loss



      

```

```python
def SVM_Model_deployement(Bid1,Biddd1):
  Bid12=Bid1.copy()
  Bid12.dropna(how='any',inplace=True)
  Bidd1=Biddd1.copy()
  Bidd1=pd.DataFrame(Bidd1.iloc[-1]).T
  pipeline_steps = [('scaler', StandardScaler()), ('svc', SVC())]

  pipeline = Pipeline(pipeline_steps)

  hyperparameters = {'svc__C': [10, 100, 1000, 10000],'svc__gamma': [1e-2, 1e-1, 1e0],'svc__kernel': ['rbf'] }
  
  clf = RandomizedSearchCV(pipeline, hyperparameters, cv=TimeSeriesSplit(n_splits=2))
  y=Bid12['Signal']

  X=Bid12.drop(['h','l','c','o','complete','volume','DayHigh','DayLow','BB_L','BB_U','Signal'],axis=1)


  Bidd1.drop(['high','low','close','open','DayHigh','DayLow','BB_L','BB_U'],axis=1,inplace=True)
  

  clf.fit(X, y)

  C = clf.best_params_['svc__C']

  kernel = clf.best_params_['svc__kernel']

  gamma = clf.best_params_['svc__gamma']

  cls = SVC(C=C, kernel=kernel, gamma=gamma)

  scaler = StandardScaler()

  cls.fit(scaler.fit_transform(X), y)

  y_predict = cls.predict(scaler.transform(Bidd1.iloc[[-1]]))
  return y_predict



  
```

```python
def Signal_d(Bid1):
  Bid1['Signal_B']=0
  Bid1['Signal_SMA']=0
  Bid1['Signal_A']=0
  Bid1['Signal_R']=0
  Bid1['Signal_M']=0
  Bid1.loc[Bid1.BB_U<=Bid1.close,'Signal_B']=-1
  Bid1.loc[Bid1.BB_L>=Bid1.close,'Signal_B']=1
  Bid1['Signal_B']=Bid1.Signal_B.map(lambda x: x if (x==1 or x==-1)  else 0)
  Bid1.loc[((Bid1['SMA_10'].shift(1)<Bid1['SMA_30'].shift(1)) & (Bid1['SMA_10']>Bid1['SMA_30'])), 'Signal_SMA' ]=1
  Bid1.loc[((Bid1['SMA_10'].shift(1)>Bid1['SMA_30'].shift(1)) & (Bid1['SMA_10']<Bid1['SMA_30'])), 'Signal_SMA' ]=-1
  
  Bid1['Signal_SMA']=Bid1.Signal_SMA.map(lambda x: x if (x==1 or x==-1) else 0)
  Bid1.loc[(Bid1.Rsi>=70),'Signal_R']=1
  Bid1.loc[(Bid1.Rsi<=30),'Signal_R']=-1
  
  Bid1['Signal_R']=Bid1.Signal_R.map(lambda x: x if (x==1 or x==-1) else 0)
  
  if len(Bid1)>=28:
    Bid1.loc[(Bid1.ADX>=25) & (Bid1.ADX_pos<=Bid1.ADX_neg),'Signal_A']=-1

    Bid1.loc[(Bid1.ADX>=25) & (Bid1.ADX_pos>=Bid1.ADX_neg),'Signal_A']=1
    Bid1['Signal_A']=Bid1.Signal_A.map(lambda x: x if (x==1 or x==-1) else 0)

  Bid1.loc[(Bid1['MACD'].shift(1)<Bid1['MACD_Signal'].shift(1)) & (Bid1['MACD']>Bid1['MACD_Signal']), 'Signal_M' ]=1
  
  Bid1.loc[(Bid1['MACD'].shift(1)>Bid1['MACD_Signal'].shift(1)) & (Bid1['MACD']<Bid1['MACD_Signal']), 'Signal_M' ]=-1
  Bid1['Signal_M']=Bid1.Signal_M.map(lambda x: x if (x==1 or x==-1) else 0)

  if "ADX" not in Bid1.columns:
    Bid1['ADX']=0
    Bid1['ADX_pos']=0
    Bid1['ADX_neg']=0


```

```python
def ADXd(Bid1):
  Bid1['ADX'],Bid1['ADX_pos'],Bid1['ADX_neg']=ta.trend.adx(Bid1.high,Bid1.low,Bid1.close),ta.trend.adx_pos(Bid1.high,Bid1.low,Bid1.close),ta.trend.adx_neg(Bid1.high,Bid1.low,Bid1.close)

def RSId(Bid1):  
  Bid1['Rsi']=ta.momentum.rsi(Bid1.close)

def MACDd(Bid1):
  Bid1['MACD'],Bid1['MACD_Signal'],Bid1['MACD_hist']=ta.trend.MACD(Bid1.close).macd(),ta.trend.MACD(Bid1.close).macd_signal(),ta.trend.MACD(Bid1.close).macd_diff()

def MAd(Bid1):
  Bid1['SMA_10'],Bid1['SMA_30']=Bid1['close'].rolling(window=10).mean(),Bid1['close'].rolling(window=30).mean()

def BBandsd(Bid1):
  Bid1['BB_U'],Bid1['BB_L']=ta.volatility.bollinger_hband(Bid1.close,window_dev=2.3),ta.volatility.bollinger_lband(Bid1.close,window_dev=2.3)

def Indicatorsd(Bid1):
  RSId(Bid1)
  MACDd(Bid1)
  # ADXd(Bid1)
#   Bid1['ADX'],Bid1['ADX_pos'],Bid1['ADX_neg']=0
  MAd(Bid1)
  BBandsd(Bid1)
  

```

```python
def jsontodataframe(json):
    file = json.json()
    price = file['candles']
    time = []
    close =[]
    high=[]
    openp=[]
    low=[]
    for cndle in price:
        time.append(cndle['time'])
        close.append(float(cndle['mid']['c']))
        high.append(float(cndle['mid']['h']))
        openp.append(float(cndle['mid']['o']))
        low.append(float(cndle['mid']['l']))
    dataframe = pd.DataFrame({"open":openp,"high":high,"low":low,"close":close})
    dataframe.index =pd.to_datetime(time)

    return dataframe
       
```
